```python
from ultralytics import YOLO
import os

# 1. Menentukan path ke file dataset.yaml
yaml_path = '../data/processed/dataset.yaml'

# 2. Inisialisasi Model Baseline
# Memuat arsitektur YOLOv8 versi Nano (paling ringan)
print("Memuat arsitektur YOLOv8 Nano...")
model = YOLO('yolov8n.pt')

# 3. Memulai Proses Training Baseline
print("Memulai training model baseline...")

# Parameter training dasar
results = model.train(
    data=yaml_path,
    epochs=10,             # Untuk baseline, kita coba 10 putaran dulu agar cepat
    imgsz=640,             # Mengubah ukuran gambar menjadi standar YOLO (640x640)
    batch=8,               # Memproses 8 gambar sekaligus (ubah ke 4 jika laptop terasa berat/error)
    project='../models',   # Menyimpan hasil di folder models yang sudah kita buat
    name='baseline'        # Nama sub-folder hasil training
)

print("Training baseline selesai! Hasil dan metrik evaluasi tersimpan di folder models/baseline/")
```
